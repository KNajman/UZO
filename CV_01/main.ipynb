```python
import numpy as np
import matplotlib.pyplot as plot
plot.rcParams.update({'font.size': 8})
import cv2

# read images
images = []
for i in range(1, 10):
    images.append(cv2.imread('im0'+str(i)+'.jpg'))

#convert to RGB
rgb_images = [cv2.cvtColor(image, cv2.COLOR_BGR2RGB) for image in images]

# Convert to grayscale
gray_images = [cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) for image in images]

# histogram computation and normalization
# histograms = [] 
# for image in gray_images:
#     h = cv2.calcHist([image], [0], None, [256], [0, 256])
#     cv2.normalize(h, h, 0, 255, cv2.NORM_MINMAX)
#     histograms.append(h)

# plot images by histogram similarity from the most similar to the most different
plot.figure(figsize=(15, 15))
position_in_fig = 1
# for count, value in enumerate(rgb_images):
#     plot.subplot(9, 9, position_in_fig)
#     plot.title('Image ' + str(position_in_fig))
#     plot.imshow(rgb_images[count])  
      
#     histogram_comparison = []
#     for count2, value2 in enumerate(rgb_images):
#         if count != count2:
#             histogram_comparison.append((cv2.compareHist(histograms[count], histograms[count2], cv2.HISTCMP_CORREL), count2))
    
#     histogram_comparison.sort(reverse=True)
#     for image_index in histogram_comparison:
#         position_in_fig += 1
#         plot.subplot(9, 9, position_in_fig)
#         plot.title('Image ' + str(position_in_fig))
#         plot.imshow(rgb_images[image_index[1]])
        
#     position_in_fig += 1
    
for count, value in enumerate(rgb_images):
    plot.subplot(9, 9, position_in_fig)
    plot.title('Original img' + str(position_in_fig))
    plot.imshow(rgb_images[count])  
    
    histogram_comparison = []
    for count2, value2 in enumerate(rgb_images):
        if count != count2:
            h1 = cv2.calcHist([gray_images[count]], [0], None, [256], [0, 256])
            h2 = cv2.calcHist([gray_images[count2]], [0], None, [256], [0, 256])
            cv2.normalize(h1, h1, 0, 255, cv2.NORM_MINMAX)
            cv2.normalize(h2, h2, 0, 255, cv2.NORM_MINMAX)
            histogram_comparison.append((cv2.compareHist(h1, h2, cv2.HISTCMP_CORREL), count2))
    
    histogram_comparison.sort(reverse=True)
    for image_index in histogram_comparison:
        position_in_fig += 1
        plot.subplot(9, 9, position_in_fig)
        plot.title('Similar img ' + str(position_in_fig))
        plot.imshow(rgb_images[image_index[1]])
        
    position_in_fig += 1


        
    
    
```
